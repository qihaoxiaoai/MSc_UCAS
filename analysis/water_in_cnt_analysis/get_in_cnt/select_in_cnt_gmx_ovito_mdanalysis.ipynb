```python
import MDAnalysis as mda
import numpy as np

def analyze_trajectory(data_path, outdir, radius=None):
    trjtpr = data_path + "/cnt.tpr"
    trjxtc = data_path + "/6-6-6.xtc"

    # 读取轨迹
    u = mda.Universe(trjtpr, trjxtc)

    # 选择resname为CNT且type为CA的原子
    selected_atoms = u.select_atoms('resname CNT and type CA')
    cnt_c_atoms = u.select_atoms('resname CNT and name C')
    z_mean = cnt_c_atoms.positions[:,2].mean()

    # 获取x, y坐标的最大值和最小值
    c_xmax = selected_atoms.positions[:, 0].max()
    c_xmin = selected_atoms.positions[:, 0].min()
    c_ymax = selected_atoms.positions[:, 1].max()
    c_ymin = selected_atoms.positions[:, 1].min()
    c_zmax = selected_atoms.positions[:, 2].max()
    c_zmin = selected_atoms.positions[:, 2].min()
    # 如果radius没有被设置，计算它
    if radius is None:
        radius = max(c_xmax - c_xmin, c_ymax - c_ymin) / 2

    # 计算四个坐标
    sqrt2 = np.sqrt(2)
    ld = (c_xmin + (radius - radius/sqrt2), c_ymin + (radius - radius/sqrt2))
    rd = (c_xmin + (radius + radius/sqrt2), c_ymin + (radius - radius/sqrt2))
    lu = (c_xmin + (radius - radius/sqrt2), c_ymin + (radius + radius/sqrt2))
    ru = (c_xmin + (radius + radius/sqrt2), c_ymin + (radius + radius/sqrt2))

    # 计算k1, k2, k3, k4
    k1 = ld[1] + ld[0]-0.2
    k2 = rd[1] - rd[0]
    k3 = lu[1] - lu[0]
    k4 = ru[1] + ru[0]

    # 输出结果
    results = {
        "radius": radius,
        "c_xmax": c_xmax,
        "c_xmin": c_xmin,
        "c_ymax": c_ymax,
        "c_ymin": c_ymin,
        "c_zmax": c_zmax,
        "c_zmin": c_zmin,
        "ld": ld,
        "rd": rd,
        "lu": lu,
        "ru": ru,
        "outside_condition": f"(Position.X < {c_xmin} || Position.X > {c_xmax} || Position.Y < {c_ymin} || Position.Y > {c_ymax}) || (Position.Y + Position.X < {k1}) || (Position.Y - Position.X < {k2}) || (Position.Y - Position.X > {k3}) || (Position.Y + Position.X > {k4})",
        "inside_condition": f"(Position.X >= {c_xmin} && Position.X <= {c_xmax} && Position.Y >= {c_ymin} && Position.Y <= {c_ymax}) && (Position.Y + Position.X >= {k1} && Position.Y - Position.X >= {k2} && Position.Y - Position.X <= {k3} && Position.Y + Position.X <= {k4})",
        "gmx_outside_condition": f"(x < {c_xmin/10} or x > {c_xmax/10} or y < {c_ymin/10} or y > {c_ymax/10}) or (y + x < {k1/10}) or (y - x < {k2/10}) or (y - x > {k3/10}) or (y + x > {k4/10})",
        "gmx_inside_condition": "group \"Water\" and within 0.11 of (name OW and (" + f"(x >= {c_xmin/10} and x <= {c_xmax/10} and y >= {c_ymin/10} and y <= {c_ymax/10}) and (y + x >= {k1/10} and y - x >= {k2/10} and y - x <= {k3/10} and y + x <= {k4/10})" + "))"
    }

    return results

# 使用函数
data_path = '/home/jxhe/works/toolkits/01-fixdp-6-12-6-19/6-6-6/'
outdir = '/home/jxhe/works/toolkits/01-fixdp-6-12-6-19/6-6-6/'
results = analyze_trajectory(data_path, outdir)
for key, value in results.items():
    print(f"{key}: {value}")
```
